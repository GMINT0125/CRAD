```python
#Augmentation Train image for CRAD
import sys
import os
import json
import cv2

data_path = "../data/cropped_data"
json_train = "../data/json/train.json"
aug_json_train = "../data/json/train_aug.json"

images = os.listdir(data_path)
data = []
aug_num = 3
window = 150

with open(json_train, 'r', encoding='utf-8') as f, open(aug_json_train, 'w', encoding='utf-8') as f2: 
    for line in f:
        info = json.loads(line.strip())
        f2.write(line) 

        filename = info['filename']
        label = info['label']
        label_name = info['label_name']
        cls_name = info['clsname']


        image_path = os.path.join(data_path, filename)

        img = cv2.imread(image_path, cv2.IMREAD_COLOR)
        height, width, _ = img.shape

        cnt = 0 
        flag = False
        #augmentation
        if width > height: #증강 x
            continue
        else: #증강 O
            for i in range(aug_num):
                if cnt + width > height:
                    patch = img[(height - width):height, 0:width]
                    flag = True
                else:   
                    patch = img[cnt:cnt + width, 0:width]
                    cnt += window
                
                aug_filename = filename.split(".")[0] + "_aug" + str(i) + ".jpg"
                aug_path = os.path.join(data_path, aug_filename)
                cv2.imwrite(aug_path, patch)

                aug_info = {
                    "filename": aug_filename,
                    "label": label,
                    "label_name": label_name,
                    "clsname": cls_name
                }
                f2.write(json.dumps(aug_info, ensure_ascii=False) + '\n')
                if flag:
                    break
       
```
